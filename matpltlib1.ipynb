```python
df1=pd.read_csv('batsman_season_record.csv')
df1
```

```python
plt.bar(df1['batsman'],df1['2015'],width=0.5)
# plt.bar(df1['batsman'],df1['2016'])
# plt.bar(df1['batsman'],df1['2017'])
```

```python
# multiple bar chart
plt.bar(np.arange(df1.shape[0]) - 0.2 ,df1['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df1.shape[0]) ,df1['2016'],width=0.2)
plt.bar(np.arange(df1.shape[0]) + 0.2 ,df1['2017'],width=0.2)

plt.xticks(np.arange(df1.shape[0]),df1['batsman'])
```

```python
# when we have long categories names 
children=[10,22,30,23]
colors=['red red red red','pink pink pink pink','blue blue bblue blue','green']
plt.bar(colors,children,color='green')

plt.xticks(rotation='vertical')
```

```python
# stack bar chart
df1
plt.bar(df1['batsman'],df1['2017'])
plt.bar(df1['batsman'],df1['2016'],bottom=df1['2017'],)
plt.bar(df1['batsman'],df1['2015'],bottom=df1['2016']+df1['2017'])
```

```python
plt.bar(df1['batsman'],df1['2017'],label='2017')
plt.bar(df1['batsman'],df1['2016'],bottom=df1['2017'],label='2016')
plt.bar(df1['batsman'],df1['2015'],bottom=df1['2016']+df1['2017'],label='2015')

plt.legend()
```
