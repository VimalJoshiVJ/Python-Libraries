```python
data= [12,15,20,23,32,50]
plt.hist(data,bins=[10,20,30,40,50,60]) # bins we can give by our choice we can give any data between which our data lies 
```

```python
data= [12,15,20,23,32,50]
plt.hist(data,bins=[10,20,30,40,50,60])
plt.show()
```

```python
data= [20,19,15,16]
subject=["hindi","eng","sciee","maths"]
plt.pie(data,labels=subject)
plt.show()
```

```python
# if we need percentage we use autopct= '%0.1f%%' function
```

```python
data= [20,19,15,16]
subject=["hindi","eng","sciee","maths"]
plt.pie(data,labels=subject,autopct= '%0.1f%%')
plt.show()
```

```python
# if we need shadow outside we use explode function
```

```python
data= [20,19,15,16]
subject=["hindi","eng","sciee","maths"]
plt.pie(data,labels=subject,autopct= '%0.1f%%',explode=[0.2,0.1,0,0]) #values we can give by our own
plt.show()
```
